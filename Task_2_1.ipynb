```python
temp_input = input("Enter the current temperature in Celsius: ")
temp = float (temp_input)

# Check the temperature range
if temp >= 30 and temp <= 50:
    print("It's a hot day. Stay hydrated!")
elif temp >= 20 and temp < 30:
    print("It's a warm day. Enjoy the weather!")
elif temp >= 10 and temp < 20:
    print("It's a cool day. Wear a jacket!")
elif temp < 10 and temp >= -15:
    print("It's a cold day. Stay warm!")
elif temp > 50 or temp < -15:
    print("Enter a Valid Temprature")
```
