```python
students = [
    {"name": "Amor", "grades": [85, 65, 88]},
    {"name": "Omar", "grades": [70, 75, 72]},
    {"name": "Mohamed", "grades": [82, 78, 95]}
]

print("Student Average Grades:\n")

for student in students:
    name = student["name"]
    grades = student["grades"]
    average = sum(grades) / len(grades)

    print(f"{name}: Average Grade = {average:.2f}")
```
