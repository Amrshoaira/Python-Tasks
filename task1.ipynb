```python
#variables for the prices of three items
meat = 55
fish = 45
fruits = 25

#variable for the total budget
total_budget = 130

#total cost of the items
total_cost = meat + fish + fruits

#Compare the total cost with the budget
print("Total Cost:", total_cost)
print("Budget:", total_budget)

#the difference
difference = total_budget - total_cost

if total_cost > total_budget:
    print("Over budget by:", abs(difference))
elif total_cost < total_budget:
    print("Under budget. Money left:", difference)
else:
    print("Total cost = budget")
```
